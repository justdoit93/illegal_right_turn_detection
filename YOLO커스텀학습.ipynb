```python
from ultralytics import YOLO
import cv2
import torch
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

### 커스텀 학습 코드

```python
# 모델 설정 및 학습 // 경로 주의
# 모델 설정
model = YOLO("yolov8s.pt") 

# 모델 학습 // roboflow에서 다운받은 yaml 파일 이름은 'data.yaml' >> 임의로 변경하였음
model.train(data='./motorcycle_rider3.yaml' , epochs=29, name='8se29_v3', ) # name은 모델이 저장될 파일 이름, drop_out=0.0(default) >> 과적합 관리
```

```python
# 학습된 모델의 클래스 확인
model.names
```

### 커스텀 학습된 모델 확인 코드

```python
# 결과 확인(우측 보도)
# model = YOLO('./runs/detect/8le100/weights/best.pt') # 학습된 가중치 불러오기

path_video = './test1.mp4' # 영상 경로

cap = cv2.VideoCapture(path_video)

while cap.isOpened():
    success, frame = cap.read()

    if success:
        results = model(frame)

        annotated_frame = results[0].plot()

        cv2.imshow("result", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
model = YOLO('./runs/detect/8se20_v/weights/best.pt') # 학습된 가중치 불러오기


path_video = './test1.mp4' # 영상 경로
output_video_path = './8se24_v3_right.mp4' # 출력 경로

cap = cv2.VideoCapture(path_video)

# VideoWriter를 초기화합니다. 출력 파일 이름, FourCC 코드, FPS, 프레임 크기를 설정합니다.
fourcc = cv2.VideoWriter_fourcc(*'mp4v') # 코덱을 설정합니다.
fps = cap.get(cv2.CAP_PROP_FPS) # 입력 영상의 FPS를 얻습니다.
frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))) # 프레임의 크기를 얻습니다.

out = cv2.VideoWriter(output_video_path, fourcc, fps, frame_size)

while cap.isOpened():
    success, frame = cap.read()

    if success:
        results = model(frame)

        annotated_frame = results[0].plot()

        out.write(annotated_frame) # annotated_frame을 비디오 파일에 씁니다.

        cv2.imshow("result", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

cap.release()
out.release() # 영상 파일을 닫고 메모리를 해제합니다.
cv2.destroyAllWindows()
```
